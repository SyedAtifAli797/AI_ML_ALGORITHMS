```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Shopping_Mart.csv')
```

```python
data = df.groupby(['InvoiceNo', 'Description'])['Quantity'].sum()
data
```

```python
data = data.unstack().reset_index()
```

```python
data
```

```python
data = data.fillna(0).set_index('InvoiceNo')
data.head()
```

For any transaction, if the quantity of an item is >=1 then its encoded as 1 (bought), else 0 (not bought)

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
```

```python
data = data.applymap(encode_units)
data
```

As we want frequently occuring items, let's ignore all the transactions where only one product is bought

```python
data = data[(data > 0).sum(axis=1) >= 2]
data
```

### Apriori Algorithm

```python
from mlxtend.frequent_patterns import apriori
```

```python
frequent_itemset = apriori(data, min_support=0.03, use_colnames=True)
```

```python
frequent_itemset
```

```python
frequent_itemset = frequent_itemset.sort_values('support', ascending=False).reset_index(drop=True)
```

```python
frequent_itemset
```

```python
frequent_itemset['length'] = frequent_itemset['itemsets'].apply(lambda x: len(x))
```

```python
frequent_itemset
```

### Association Rules

- Lift helps eliminate rules that occur due to chance or high popularity of items.
- Lift emphasizes how much the presence of one item increases the likelihood of another.
- A high lift value indicates a strong association, making it valuable for targeting strategies (e.g., marketing).

Due to all these factors let us find the association between the items based on the lift

```python
from mlxtend.frequent_patterns import association_rules
```

```python
association = association_rules(frequent_itemset, metric='lift', min_threshold=1, num_itemsets=len(data))
```

```python
association.head()
```

```python
association.sort_values('lift', ascending=False, inplace=True)
```

```python
association.reset_index(drop=True, inplace=True)
```

```python
association.head(5)
```

 - We can see GREEN REGENCY TEACUP AND SAUCER and PINK REGENCY TEACUP AND SAUCER are the items that has the highest association since these two items has the highest lift value, i.e 14.6
 
- This tells us that 'PINK REGENCY TEACUP AND SAUCER' is 14.6 times more likely to be bought by the customers who buy 'GREEN REGENCY TEACUP AND SAUCER' compared to the default likelihood sale of 'PINK REGENCY TEACUP AND SAUCER’

- Since the antecedent support > consequent support, the rule that applies is {GREEN REGENCY TEACUP AND SAUCER -> PINK REGENCY TEACUP AND SAUCER}. This means that a customer has a higher tendency to buy PINK REGENCY TEACUP AND SAUCER AFTER they buy GREEN REGENCY TEACUP AND SAUCER. Not the other way around.

- The confidence level for the rule is 0.6177, which shows that out of all the transactions that contain both “GREEN REGENCY TEACUP AND SAUCER”, ~62% contain PINK REGENCY TEACUP AND SAUCER too

Similarly, we can draw other business conclusions from these association rules, which will lead in increase in revenue!
